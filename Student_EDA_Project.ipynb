```python
avg_scores = df.groupby('gender')['math score'].mean()
plt.figure(figsize=(6, 6))
plt.pie(avg_scores, labels=avg_scores.index, autopct='%1.1f%%', startangle=90)
plt.title('Average Math Score by Gender')
plt.axis('equal')
plt.show()
```

```python
sns.barplot(x=df['gender'],y=df['math score'], data=df)
plt.title("Gender by math score")
plt.show()
```

```python
sns.histplot(df['math score'],kde=True)
plt.title("Maths by  distibutions")
plt.show()
```
